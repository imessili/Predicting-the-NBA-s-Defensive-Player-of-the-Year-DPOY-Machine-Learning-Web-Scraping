```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python

# Liste des fichiers CSV
players = ['players_1993_1998.csv','players_1999_2004.csv','players_2005_2010.csv','players_2011_2016.csv','players_2017_2023.csv'] # Ajoutez tous vos fichiers ici

# Colonnes à supprimer
columns_to_drop = ['Season', 'Year', 'Player_name', 'won_dpoy_last_season',
                   'Max_points', 'Tm', 'Lg', 'Pos', 'Rank', 'Arena_misc',
                   'Attendance_misc']

for file in players:
    # Charger le DataFrame
    all_players = pd.read_csv(file)

    # Supprimer les colonnes indésirables
    all_players.drop(columns=columns_to_drop, inplace=True)

    # Remplacer '--' par NaN et convertir les colonnes en numériques
    all_players.replace('--', np.nan, inplace=True)
    all_players = all_players.apply(pd.to_numeric, errors='coerce')
    #remplacer par zero
    all_players.fillna(0, inplace=True)
    

    # Séparer notre DataFrame
    X = all_players.drop('Points_won', axis=1)  # Caractéristiques
    Y = all_players['Points_won']  # Cible

    # Calculer les scores d'information mutuelle pour chaque caractéristique
    mi_scores = mutual_info_classif(X, Y)

    # Créer un DataFrame pour les scores d'information mutuelle
    mi_df = pd.DataFrame({'Feature': X.columns, 'MI_Score': mi_scores})
    mi_df = mi_df.sort_values(by='MI_Score', ascending=True)

    # Supprimer les colonnes non nommées, si elles existent
    mi_df = mi_df.drop(columns=mi_df.columns[mi_df.columns.str.contains('Unnamed')])
    #créer fichier
    mi_df.to_csv(f'mutual_information_scores_{file}.csv', index=False)
    

    # Tracer les scores d'information mutuelle
    plt.figure(figsize=(15, 25))
    plt.barh(mi_df['Feature'], mi_df['MI_Score'], align="center")
    plt.xlabel('Mutual Information Score')
    plt.ylabel('Features')
    plt.title(f'Mutual Information Scores of Features for {file}')
    plt.savefig(f'mutual_information_{file}.png')
    plt.show()
```

```python
selector = SelectFromModel(estimator=RandomForestClassifier())
model = RandomForestClassifier()
players = ['players_1993_1998.csv','players_1999_2004.csv','players_2005_2010.csv','players_2011_2016.csv','players_2017_2023.csv'] # Ajoutez tous vos fichiers ici
# Colonnes à supprimer
columns_to_drop = ['Season', 'Year', 'Player_name', 'won_dpoy_last_season',
                   'Max_points', 'Tm', 'Lg', 'Pos', 'Rank', 'Arena_misc',
                   'Attendance_misc']
for file in players:
    # Charger le DataFrame
    players = pd.read_csv(file)

    # Supprimer les colonnes indésirables
    players.drop(columns=columns_to_drop, inplace=True)

    # Remplacer '--' par NaN et convertir les colonnes en numériques
    players.replace('--', np.nan, inplace=True)
    players = players.apply(pd.to_numeric, errors='coerce')
    #remplacer par zero
    players.fillna(0, inplace=True)
    
    # Séparer notre DataFrame
    X = players.drop('Points_won', axis=1)  # Caractéristiques
    Y = players['Points_won']  # Cible

    # Ré-entraînement du modèle SelectFromModel 
    selector.fit(X, Y)

    # Obtention des caractéristiques importantes sans 'First_votes'
    important = X.columns[selector.get_support()]

    # Ré-entraînement du modèle pour obtenir les importances des caractéristiques 
    model.fit(X, Y)
    importances = model.feature_importances_

    # Obtenir les indices des caractéristiques importantes
    indices = np.argsort(importances)[-50:] 
    
    # Créer un DataFrame pour les caractéristiques importantes et leurs scores
    features_importances_df = pd.DataFrame({'Feature': X.columns, 'Importance': importances})
    features_importances_df = features_importances_df.sort_values(by='Importance', ascending=False)  # Tri par importance

    # Enregistrer le DataFrame des caractéristiques importantes dans un fichier CSV
    importance_csv_filename = f'feature_importances_{file.split(".")[0]}.csv'
    features_importances_df.to_csv(importance_csv_filename, index=False)
   
    # Création du graphique des importances des caractéristiques 
    plt.figure(figsize=(10, 6))
    plt.title(f'caractéristiques importantes dans {file}')
    plt.barh(range(len(indices)), importances[indices], color='green', align='center')
    plt.yticks(range(len(indices)), [X.columns[i] for i in indices])
    plt.xlabel('Importance relative')
    plt.show()
    
    
```

```python
selector = SelectFromModel(estimator=LogisticRegression(max_iter=1000))
model = LogisticRegression(max_iter=5000)
players = ['players_1993_1998.csv','players_1999_2004.csv','players_2005_2010.csv','players_2011_2016.csv','players_2017_2023.csv'] # Ajoutez tous vos fichiers ici
# Colonnes à supprimer
columns_to_drop = ['Season', 'Year', 'Player_name', 'won_dpoy_last_season',
                   'Max_points', 'Tm', 'Lg', 'Pos', 'Rank', 'Arena_misc',
                   'Attendance_misc']
for file in players:
    # Charger le DataFrame
    players = pd.read_csv(file)

    # Supprimer les colonnes indésirables
    players.drop(columns=columns_to_drop, inplace=True)

    # Remplacer '--' par NaN et convertir les colonnes en numériques
    players.replace('--', np.nan, inplace=True)
    players = players.apply(pd.to_numeric, errors='coerce')
    players.fillna(0, inplace=True)
    
    # Séparer notre DataFrame
    X = players.drop('Points_won', axis=1)  # Caractéristiques
    Y = players['Points_won']  # Cible

    # Ré-entraînement du modèle SelectFromModel avec LogisticRegression
    selector.fit(X, Y)

    # Obtention des caractéristiques importantes avec LogisticRegression
    important_features = X.columns[selector.get_support()]

    # Ré-entraînement du modèle pour obtenir les coefficients (importances) avec LogisticRegression
    model.fit(X, Y)
    importances = np.abs(model.coef_[0])

    # Créer un DataFrame pour les caractéristiques importantes et leurs importances
    features_importances_df = pd.DataFrame({'Feature': X.columns, 'Importance': importances})
    features_importances_df = features_importances_df.sort_values(by='Importance', ascending=False)  # Tri par importance

    # Enregistrer le DataFrame des caractéristiques importantes dans un fichier CSV
    importance_csv_filename = f'logistic_regression_feature_importances_{file.split(".")[0]}.csv'
    features_importances_df.to_csv(importance_csv_filename, index=False)

    # Obtenir les indices des caractéristiques importantes
    indices = np.argsort(importances)[-40:]  

    # Création du graphique des importances des caractéristiques
    plt.figure(figsize=(10, 6))
    plt.title(f'importantes avec LogisticRegression dans {file}')
    plt.barh(range(len(indices)), importances[indices], color='blue', align='center')
    plt.yticks(range(len(indices)), [X.columns[i] for i in indices])
    plt.xlabel('Importance relative')
    plt.show()
```
