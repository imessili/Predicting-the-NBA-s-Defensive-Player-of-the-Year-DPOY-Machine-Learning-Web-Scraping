```python
from bs4 import BeautifulSoup
import pandas as pd
## reading the html page
f = open("bball.html",encoding="utf-8")
page = f.read()

## saving the DPOY table
soup = BeautifulSoup(page,'html.parser')
##soup.find('tr', class_="over_header").decompose()
full_table = soup.find_all(id="DataTables_Table_0")[0]

column_headers = ['Season', 'Player', 'Team(s)', 'Minutes',
                  'Offense Position', 'Offensive Archetype',
                  'LEBRON WAR', 'LEBRON', 'O-LEBRON', 'D-LEBRON']
    
rows = []
for row in full_table.select('tbody tr'):
    row_data = [td.get_text(strip=True) for td in row.find_all(['td', 'th'])]
    rows.append(row_data[1:])
    
full_table = pd.DataFrame(rows, columns=column_headers)
full_table.reset_index(drop=True, inplace=True)
    
```
