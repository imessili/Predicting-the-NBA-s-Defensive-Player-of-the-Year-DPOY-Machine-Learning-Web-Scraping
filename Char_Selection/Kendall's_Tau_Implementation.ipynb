```python
import os
import pandas as pd
from scipy.stats import kendalltau

# Define the main directory
main_dir = r'C:\Users\Abdou\Desktop\Projet_Prediction_DPOY_2023_2024\Char_Selection'

# Define the subdirectories
subdirectories = [
    '1993_1997', '1998_2002', '2003_2007', 
    '2008_2012', '2013_2017', '2018_2023', 
    'All_years_1993_2023'
]

# Function to compare two DataFrames using Kendall's tau coefficient
def compare_dataframes(df1, df2):
    # Extract only the relevant columns for comparison
    data1 = df1['Feature'].rank().values
    data2 = df2['Feature'].rank().values
    
    # Compute Kendall's tau coefficient
    tau, _ = kendalltau(data1, data2)
    return tau

# Create an empty DataFrame to store the results
results_RF = pd.DataFrame(index=subdirectories, columns=subdirectories)
results_XGB = pd.DataFrame(index=subdirectories, columns=subdirectories)
results_RF_XGB = pd.DataFrame(index=subdirectories, columns=subdirectories)

# Iterate through each subdirectory
for subdir1 in subdirectories:
    # Check if subdir1 contains a file with "RF" in its name
    files1_RF = [file for file in os.listdir(os.path.join(main_dir, subdir1)) 
                 if "RF" in file and file.endswith('.csv')]
    files1_XGB = [file for file in os.listdir(os.path.join(main_dir, subdir1)) 
                  if "XGB" in file and file.endswith('.csv')]
    if files1_RF:
        for subdir2 in subdirectories:
            # Check if subdir2 also contains a file with "RF" in its name
            files2_RF = [file for file in os.listdir(os.path.join(main_dir, subdir2)) 
                         if "RF" in file and file.endswith('.csv')]
            if files2_RF:
                # Read the CSV files into pandas DataFrames
                df1_RF = pd.read_csv(os.path.join(main_dir, subdir1, files1_RF[0]))
                df2_RF = pd.read_csv(os.path.join(main_dir, subdir2, files2_RF[0]))
                
                # Compare the DataFrames and store the result
                tau = compare_dataframes(df1_RF, df2_RF)
                if tau is not None:
                    results_RF.loc[subdir1, subdir2] = tau

    if files1_XGB:
        for subdir2 in subdirectories:
            # Check if subdir2 also contains a file with "XGB" in its name
            files2_XGB = [file for file in os.listdir(os.path.join(main_dir, subdir2)) 
                          if "XGB" in file and file.endswith('.csv')]
            if files2_XGB:
                # Read the CSV files into pandas DataFrames
                df1_XGB = pd.read_csv(os.path.join(main_dir, subdir1, files1_XGB[0]))
                df2_XGB = pd.read_csv(os.path.join(main_dir, subdir2, files2_XGB[0]))
                
                # Compare the DataFrames and store the result
                tau = compare_dataframes(df1_XGB, df2_XGB)
                if tau is not None:
                    results_XGB.loc[subdir1, subdir2] = tau

            if files1_RF:
                # Read the CSV files into pandas DataFrames for cross comparison
                df1_RF = pd.read_csv(os.path.join(main_dir, subdir1, files1_RF[0]))
                
                # Compare the DataFrames and store the result
                tau = compare_dataframes(df1_RF, df1_XGB)
                if tau is not None:
                    results_RF_XGB.loc[subdir1, subdir2] = tau

# Display the results DataFrames
print("Kendall's Tau Results for RandomForestRegressor and RandomForestRegressor:")
display(results_RF)
print("\nKendall's Tau Results for XGBoostingRegressor and XGBoostingRegressor:")
display(results_XGB)
print("\nKendall's Tau Results for RandomForestRegressor and XGBoostingRegressor:")
display(results_RF_XGB)
```
