```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
file_path = 'top_N_features_Mutual_Information_ScoreAll.csv'
df = pd.read_csv(file_path)

# Create a bar plot
plt.figure(figsize=(12, 25))
plt.barh(df['Feature'], df['MI_Score'], color='blue')
plt.xlabel('Mutual Information Score')
plt.ylabel('Feature')
plt.title('Top N Features by Mutual Information Score')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest score at the top
# Save the plot as a PNG file
plt.savefig('top_N_features_MI_Score.png', bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
file_path = 'top_N_features_PersonsRAll.csv'
df = pd.read_csv(file_path)

# Create a bar plot for correlation
plt.figure(figsize=(12, 25))
plt.barh(df['Feature'], df['correlation'], color='skyblue')
plt.xlabel('Correlation')
plt.ylabel('Feature')
plt.title('Top N Features by Correlation (PersonsRAll)')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest score at the top

# Save the plot as a PNG file
plt.savefig('top_N_features_PersonsRAll.png', bbox_inches='tight')

# Show the plot
plt.show()
```
