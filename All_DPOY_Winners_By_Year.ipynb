```python
## All years from 1993 to 2023
years = list(range(1993,2024))
```

```python
## url to all DPOY winners
url_voting_awards = "https://www.basketball-reference.com/awards/awards_{}.html"
```

```python
## to execute webChromeDriver
## allows developers to automate interactions with Chrome
## opening web pages, filling forms, clicking buttons, and extracting information
import os
## Js framework for automating web browsers
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

# Create a new instance of the Chrome driver
driver = webdriver.Chrome(
    executable_path="./chromedriver.exe"
    )
```

```python
## extracting all the DPOY pages by year
for year in years:
    url = url_voting_awards.format(year)

    ## scrolling to the bottum of the page
    driver.get(url)
    driver.execute_script("window.scrollTo(1,10000)")

    ## pausing 5sec to avoid website protection
    time.sleep(5)

    ## save html file
    with open("Voting_Awards/{}.html".format(year), "w+", encoding="utf-8") as f:
        f.write(driver.page_source)
```

```python
from bs4 import BeautifulSoup
import pandas as pd
## extracting DPOY tables by year
for year in years:
    ## opening html pages
    with open("Voting_Awards/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()
    ## extraction    
    soup = BeautifulSoup(page, 'html.parser')
    all_dpoy_by_year_table = soup.find_all(id="dpoy")[0]
    all_dpoy_by_year = pd.read_html(str(all_dpoy_by_year_table))[0]
    all_dpoy_by_year.to_csv("Voting_Awards/CSVs/{}.csv".format(year))
```
